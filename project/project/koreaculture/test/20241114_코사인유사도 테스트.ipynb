```python
db1_df = pd.read_csv("../data/국민여가활동조사_2022_데이터2.csv")

db1_df
```

```python
import numpy as np
import pandas as pd
from numpy import dot
from numpy.linalg import norm

# 코사인 유사도 함수 정의
def cos_sim(A, B):
    return dot(A, B) / (norm(A) * norm(B))

# ID 열을 제외한 나머지 데이터를 벡터로 사용
data_vectors = db1_df.iloc[:5000, 1:].values  # ID를 제외한 데이터

# # 예시: 첫 번째와 두 번째 행의 코사인 유사도 계산
# similarity_1_2 = cos_sim(data_vectors[0], data_vectors[1])
# print(f'문서 1과 문서 2의 유사도: {similarity_1_2}')

# # 예시: 첫 번째와 세 번째 행의 코사인 유사도 계산
# similarity_1_3 = cos_sim(data_vectors[0], data_vectors[2])
# print(f'문서 1과 문서 3의 유사도: {similarity_1_3}')

# 전체 유사도 행렬 계산 (모든 문서 쌍에 대해)
similarity_matrix = np.zeros((len(data_vectors), len(data_vectors)))
similarity_matrix = np.zeros((5000, 5000))

for i in range(len(data_vectors)):
    # for j in range(i + 1, len(data_vectors)):  # 대칭 행렬이므로 절반만 계산
    for j in range(i + 1, 5000):  # 대칭 행렬이므로 절반만 계산
        similarity_matrix[i, j] = cos_sim(data_vectors[i], data_vectors[j])
        similarity_matrix[j, i] = similarity_matrix[i, j]

# 유사도 행렬을 데이터프레임으로 출력
# similarity_df = pd.DataFrame(similarity_matrix, index=db1_df['ID'], columns=db1_df['ID'])
similarity_df = pd.DataFrame(similarity_matrix, index=db1_df['ID'][:5000], columns=db1_df['ID'][:5000])
print(similarity_df)
```

```python
import numpy as np
import pandas as pd
from numpy import dot
from numpy.linalg import norm

# 코사인 유사도 함수 정의
def cos_sim(A, B):
    return dot(A, B) / (norm(A) * norm(B))

# 데이터 로드 및 5000개 샘플링 (예시)
data_vectors = db1_df.iloc[:5000, 1:].values  # ID를 제외한 데이터
data_ids = db1_df['ID'][:5000].values  # 5000개의 ID 추출

# # 사용자가 입력한 벡터 예시 (예: 사용자가 입력한 벡터로 임의의 값을 설정)
# user_input_vector = np.array([0, 1, 0, 0, 1, 0, 0, 1, 0, 0, ...])  # 적절한 길이의 벡터로 설정

# 91차원 테스트 벡터 생성 (0 또는 1로 임의의 값 설정)
user_input_vector = np.random.randint(2, size=91)

print("테스트 벡터:", user_input_vector)

# 입력 벡터와 각 데이터 벡터 간의 유사도를 계산하여 리스트에 저장
similarities = []
for i in range(len(data_vectors)):
    similarity = cos_sim(user_input_vector, data_vectors[i])
    similarities.append((data_ids[i], similarity))  # (ID, 유사도) 형태로 저장

# 유사도에 따라 내림차순 정렬
similarities.sort(key=lambda x: x[1], reverse=True)

# 상위 N개의 유사한 문서 추출 (예: 상위 5개)
top_n = 5
top_n_similarities = similarities[:top_n]

# 출력
print("사용자 입력과 가장 유사한 상위 5개 문서:")
for doc_id, sim in top_n_similarities:
    print(f"문서 ID: {doc_id}, 유사도: {sim}")
```
